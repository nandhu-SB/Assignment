# 3.Check whether all the zones generate the same amount of sales.

```python
z1=df['Zone - A']
z2=df['Zone - B']
z3=df['Zone - C']
z4=df['Zone - D']
```

H0 : mean_z1 = mean_z2 = mean_z3 = mean_z4

H1 : Their mean is different

```python
F,p=stat.f_oneway(z1,z2,z3,z4,axis=0)
print('F-statistics : ',F)
print('p-value : ',p)
```

```python
Here p-value<0.05 Hence we reject the null hypothesis.

Therefore,amount of sales generated in the 4 zones is different.
```
