```python
def quart(data, var):                                   ##handling outliers
    q1 = np.quantile(data[var], 0.25)
    q3 = np.quantile(data[var], 0.75)
    iqr = q3 - q1
    uplim = q3 + (iqr*1.5)
    lowlim = q1 - (iqr*1.5)
    print('q1= ',q1)
    print('q3= ',q3)
    print('iqr= ',iqr)
    print('uplim = ',uplim, 'and lowlim = ',lowlim)
```

```python
for x in l:
    print(x)
    quart(df,x)
    print('')
```

```python
def outlier(var, up, low):
    for x in df[var]:
        if x>up:
            df[var].replace(x, up,inplace=True)
        elif x<low:
            df[var].replace(x, low, inplace=True)
```

```python
outlier('Malic_Acid',5.3025,-0.6174999999999997)
outlier('Ash',3.0787500000000003, 1.6887499999999998 )
outlier('Ash_Alcanity',27.950000000000003,10.749999999999998 )
outlier('Magnesium', 135.5, 59.5)
outlier('Proanthocyanins', 3.0, 0.20000000000000018)
```

# 3. Find the optimum number of clusters in each model and create the model with the optimum number of clusters.

# K-Means Clustering:-

```python
##finding optimum no.of clusters using elbow method
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)
    print(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title("The Elbow Method")
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from the above graph we can say that optimum no.of clusters is 3.
```
