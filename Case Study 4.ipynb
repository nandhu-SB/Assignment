# 1.The company wishes to clarify whether there is any increase in sales after stepping into digital marketing.

Hypothesis: H0 :There is an increase in sales after stepping into digital marketing.

H1 : There is  no increase in sales after stepping into digital marketing.


```python
x1=np.array(df['Sales_before_digital_add(in $)'])
y1=np.array(df['Sales_After_digital_add(in $)'])
```

```python
stat.ttest_ind(x1,y1,alternative='greater')
```

alpha=0.05 

here,p-value > 0.05

so,we accept null hypothesis. 

Therefore,there is an increase in sales after stepping into digital marketing..

# 2. The company needs to check whether there is any dependency between the features “Region” and “Manager”.

Hypothesis: H0 : Region and Manager are dependent features.

H1 :  Region and Manager are independent features.

```python
contigency=pd.crosstab(df['Region'],df['Manager'])
contigency
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(contigency,annot=True,cmap="YlGnBu")
```
